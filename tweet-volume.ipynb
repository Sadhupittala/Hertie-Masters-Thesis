```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
df_bernie = pd.read_csv("data/bernie-tweet-volume.csv")
df_cruz = pd.read_csv("data/cruz-tweet-volume.csv")
```

```python
def format_date(date):
    year_month_day = date.split('T')[0].split('-')
    year = int(year_month_day[0])
    month = int(year_month_day[1])
    day = int(year_month_day[2])
    return datetime(year, month, day)
```

```python
df_bernie['end'] = df_bernie['end'].apply(format_date)
df_cruz['end'] = df_cruz['end'].apply(format_date)
```

```python
plt.figure(figsize=(18, 6), dpi=80)
plt.scatter(df_bernie['end'], df_bernie['tweet_count'])
plt.show()
```

```python
plt.figure(figsize=(18, 6), dpi=80)
plt.scatter(df_cruz['end'], df_cruz['tweet_count'])
plt.show()
```

```python
plt.figure(figsize=(18, 6), dpi=80)
plt.scatter(df_cruz['end'], df_cruz['tweet_count'])
plt.scatter(df_bernie['end'], df_bernie['tweet_count'])
plt.ylabel("Tweet Count")
plt.xlabel("Date")
plt.title("Tweet Count vs Date")
plt.legend(["Ted Cruz", "Bernie Sanders"])
plt.show()
```
